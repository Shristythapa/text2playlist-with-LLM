```python
# playlist_data = []

# for data in pj:
#   name = data['name']

#   for track in data['tracks']:
#       value = {"playlistName" :name, 'trackName': track['track_name'], 'energy': track['track_features']['energy'], 'key': track['track_features']['key'], 'loudness': track['track_features']['loudness'], 'mode': track['track_features']['mode'], 'speechiness': track['track_features']['speechiness'], 'acousticness': track['track_features']['acousticness'], 'instrumentalness': track['track_features']['instrumentalness'], 'liveness': track['track_features']['liveness'], 'valence': track['track_features']['valence'], 'tempo': track['track_features']['tempo']}
#       playlist_data.append(value)
#       print(value)


playlist_data = []

for data in pj:
    name = data.get('name')  # Use .get() method to handle missing keys gracefully

    for track in data.get('tracks', []):  # Use .get() method with a default value to handle missing tracks gracefully
        try:
            value = {
                "playlistName": name if name else None,  # Assign None if name is missing
                'trackName': track.get('track_name'),  # Use .get() method to handle missing track names gracefully
                'energy': track['track_features'].get('energy'),  # Use .get() method to handle missing energy values gracefully
                'key': track['track_features'].get('key'),
                'loudness': track['track_features'].get('loudness'),
                'mode': track['track_features'].get('mode'),
                'speechiness': track['track_features'].get('speechiness'),
                'acousticness': track['track_features'].get('acousticness'),
                'instrumentalness': track['track_features'].get('instrumentalness'),
                'liveness': track['track_features'].get('liveness'),
                'valence': track['track_features'].get('valence'),
                'tempo': track['track_features'].get('tempo')
            }
        except KeyError:
            # Handle KeyError exceptions if any required keys are missing
            continue

        playlist_data.append(value)
        print(value)
```

```python
df= pd.DataFrame(playlist_data)
```

```python
# Write the DataFrame to a CSV file
df.to_csv('/content/drive/MyDrive/final_playlist.csv', index=False)

print("CSV file created successfully.")
```
